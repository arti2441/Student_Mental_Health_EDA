# Feature Engineering


## City
- Rare and invalid city entries were grouped under "Others" to reduce noise and improve data quality.
- The City column contained high cardinality along with spelling inconsistencies and invalid entries such as names and degree labels. Incorrect entries were grouped under 'Others'. To improve interpretability and reduce dimensionality, cities were further categorized into City_Tiers (Tier 1, Tier 2, Tier 3) based on urbanization level. This enabled meaningful geographic analysis while avoiding noise and overfitting.

```python
df['City'].value_counts()
```

```python
valid_cities = df['City'].value_counts()[df['City'].value_counts() > 10].index
df['City'] = df['City'].apply(lambda x: x if x in valid_cities else 'Others')
```

```python
def assign_tier(city):
  if city in ['Mumbai','Delhi','Bangalore','Chennai','Hyderabad','Kolkata','Pune','Ahmedabad']:
    return 'Tier 1'
  elif city in ['Jaipur','Lucknow','Patna','Indore','Bhopal',
    'Nagpur','Nashik','Surat','Vadodara','Rajkot',
    'Ghaziabad','Faridabad','Meerut','Kanpur',
    'Varanasi','Ludhiana','Agra','Thane',
    'Vasai-Virar','Kalyan','Visakhapatnam']:
    return 'Tier 2'
  else:
    return 'Tier 3'
```

```python
df['City_tier'] = df['City'].apply(assign_tier)
```

```python
df['City_tier'].value_counts()
```

## Sleep Duration
- Rare and ambiguous sleep duration entries labeled as 'Others' were treated as missing values and imputed using the mode to preserve data consistency.
- Sleep duration is further grouped into quality-based categories to capture health-relevant sleep patterns.

```python
df['Sleep Duration'].value_counts()
```

```python
df['Sleep Duration'] = df['Sleep Duration'].replace('Others',np.nan)
```

```python
df['Sleep Duration'] = df['Sleep Duration'].fillna(df['Sleep Duration'].mode()[0])
```

```python
df['Sleep Duration'].value_counts()
```

```python
def sleep_quality(x):
  if x in ["'Less than 5 hours'","'5-6 hours'"]:
    return 'Poor sleep'
  elif x in ["'7-8 hours'"]:
    return 'Healthy sleep'
  else:
    return 'Excess sleep'
```

```python
df['Sleep Quality'] = df['Sleep Duration'].apply(sleep_quality)
```

```python
df['Sleep Quality'].value_counts()
```

## Degree
- The Degree column contained multiple education qualifications with varying levels. To reduce dimensionality and improve interpretability, degrees were grouped into broader education-level categories such as High School, Undergraduate, Postgraduate, and Professional.

```python
degre = df['Degree'].value_counts()
degre.index
```

```python
def classify_degree(degree):
  if degree in ['B.Com', 'B.Arch', 'BCA','B.Tech','BHM', 'BSc','B.Pharm', 'BBA','BE', 'BA']:
    return 'Undergraduate'
  elif degree in ['Class 12']:
    return 'High School'
  elif degree in ['MSc','MCA','M.Tech','M.Com','MA','M.Pharm','MHM','MBA','ME']:
    return 'Postgraduate'
  elif degree in ['B.Ed','M.Ed','LLB','LLM']:
    return 'Professional'
  elif degree in ['MBBS','MD']:
    return 'Medical'
  elif degree=='PhD':
    return 'Doctorate'
  else:
    return 'Others'
```

```python
df['Degree level'] = df['Degree'].apply(classify_degree)
```

```python
df['Degree level'].value_counts()
```

## Study Hours
Study hours were categorized into five levels representing study intensity. This binning improves interpretability, reduces noise from discrete numeric values, and enables clearer analysis of study patterns in relation to academic and mental health outcomes.

```python
df['Study Hours'].value_counts()
```

```python
def classify_study_time(study):
  if study==0:
    return 'No Study'
  elif 1<=study<=3:
    return 'Very Low'
  elif 4<=study<=6:
    return 'Low'
  elif 7<=study<=9:
    return 'Moderate'
  else:
    return 'High'
```

```python
df['Study Level'] = df['Study Hours'].apply(classify_study_time)
```

```python
df['Study Level'].value_counts()
```

## CGPA
- CGPA values were grouped into performance levels to simplify interpretation and enable categorical analysis. The classification was designed to avoid overlapping or missing ranges and reflects standard academic grading thresholds.

```python
def classify_cgpa(cgpa):
  if cgpa<6:
    return 'Low'
  elif 6<=cgpa<8:
    return 'Moderate'
  else:
    return 'High'
```

```python
df['CGPA_level'] = df['CGPA'].apply(classify_cgpa)
```

```python
df['CGPA_level'].value_counts()
```
